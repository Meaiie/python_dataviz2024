````
import pandas as pd
import re

# ฟังก์ชันสำหรับลบตัวอักษรภาษาอังกฤษและสัญลักษณ์
def remove_english_and_symbols(text):
    if isinstance(text, str): # Check if 'text' is a string
        return re.sub(r'[a-zA-Z0-9!@#$%^&*()_+=\[\]{};:"\\|,.<>\/?~`]', '', text)
    else:
        return "" # Return empty string if not a string

longest_story = ("", "", 0) # initialize longest_story (title, story, length)
shortest_story = ("", "", float('inf')) # initialize shortest_story (title, story, length)

# Assuming 'stories' is the list containing the actual stories
for index, row in folk_df.iterrows(): # Iterate over rows to access both title and description
    title = row['title']
    story = row['description']

    # กรองตัวอักษรภาษาอังกฤษและสัญลักษณ์ออก
    clean_story = remove_english_and_symbols(story)

    # คำนวณความยาวของนิทาน
    story_length = len(clean_story)

    # Update longest and shortest stories
    if story_length > longest_story[2]:
        longest_story = (title, clean_story, story_length)
    if story_length < shortest_story[2]:
        shortest_story = (title, clean_story, story_length)

# แสดงผล
print("นิทานที่ยาวที่สุด:")
print("ชื่อ:", longest_story[0])
print(longest_story[1].strip())
print("\nความยาว:", longest_story[2], "ตัวอักษร")

print("\n\nนิทานที่สั้นที่สุด:")
print("ชื่อ:", shortest_story[0])
print(shortest_story[1].strip())
print("\nความยาว:", shortest_story[2], "ตัวอักษร")
````

## Quiz 9
- วนลูปนับว่ามีนิทานที่มีคำว่าแม่หรือพ่อในคอลัมน์ title กี่เรื่อง เรื่องอะไรบ้าง
- มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
- คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือพ่อมากที่สุด



````
# prompt: - วนลูปนับว่ามีนิทานที่มีคำว่าแม่หรือพ่อในคอลัมน์ title กี่เรื่อง เรื่องอะไรบ้าง
# - มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
# - คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือพ่อมากที่สุด ไม่นับค่า nan

# วนลูปนับว่ามีนิทานที่มีคำว่าแม่หรือพ่อในคอลัมน์ title กี่เรื่อง เรื่องอะไรบ้าง
count_mom_dad = 0
mom_dad_stories = []
life_count = 0
local_count = 0
creator_counts = {}

for index, row in this_data.iterrows():
  title = row['title']
  story_type = row['type']
  creator = row['creator']
  if 'แม่' in title or 'พ่อ' in title:
    count_mom_dad += 1
    mom_dad_stories.append(title)
    if story_type == 'นิทานชีวิต':
      life_count += 1
    elif story_type == 'นิทานประจำถิ่น':
      local_count += 1
    if creator and isinstance(creator, str):  # Check if creator is not NaN and is a string
      if creator in creator_counts:
        creator_counts[creator] += 1
      else:
        creator_counts[creator] = 1

print("มีนิทานที่มีคำว่าแม่หรือพ่อในคอลัมน์ title จำนวน", count_mom_dad, "เรื่อง")
print("รายชื่อนิทาน:", mom_dad_stories)

# มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
if life_count > local_count:
  print("มีคำว่าแม่หรือพ่อในนิทานชีวิตมากกว่า")
elif local_count > life_count:
  print("มีคำว่าแม่หรือพ่อในนิทานประจำถิ่นมากกว่า")
else:
  print("มีคำว่าแม่หรือพ่อในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน")

# คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือพ่อมากที่สุด ไม่นับค่า nan
if creator_counts:  # Check if creator_counts is not empty
  most_frequent_creator = max(creator_counts, key=creator_counts.get)
  print("คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือพ่อมากที่สุดคือ:", most_frequent_creator, "จำนวน", creator_counts[most_frequent_creator], "เรื่อง")
else:
  print("ไม่มีข้อมูลคนแต่งสำหรับนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่อง")
````

````
# prompt: เรียงลำดับผู้กำกับที่มี rating เฉลี่ย มากกว่า 7 โดยที่มีจำนวนหนังมากว่า 2 เรื่อง

# คำนวณค่าเฉลี่ยของ Rating สำหรับผู้กำกับแต่ละคน
average_ratings = movie_df.groupby('Director')['Rating'].mean()

# นับจำนวนหนังที่กำกับโดยผู้กำกับแต่ละคน
movie_counts = movie_df['Director'].value_counts()

# เลือกผู้กำกับที่มี rating เฉลี่ยมากกว่า 7 และมีหนังมากกว่า 2 เรื่อง
qualified_directors = average_ratings[(average_ratings > 7) & (movie_counts > 2)].index

# เรียงลำดับผู้กำกับตาม rating เฉลี่ยจากมากไปน้อย
top_directors = average_ratings[qualified_directors].sort_values(ascending=False)

# แสดงผลลัพธ์
print("ผู้กำกับที่มี rating เฉลี่ยมากกว่า 7 และมีหนังมากกว่า 2 เรื่อง (เรียงลำดับตาม rating):")
for director, rating in top_directors.items():
    num_movies = movie_df[movie_df['Director'] == director].shape[0]
    print(f"- {director}: ค่าเฉลี่ย {rating:.2f} จาก {num_movies} เรื่อง")
````

````
# prompt: เรียงลำดับผู้กำกับที่มี rating มากกว่า 7  และมีจำนวนหนังที่กำกับไม่น้อยกว่า 5 เรื่อง  โดยที่บอก genre ทั้งหมดที่เขากำกับมาด้วย(ไม่แสดงผล genre ที่ซ้ำ)

# คำนวณค่าเฉลี่ยของ Rating สำหรับผู้กำกับแต่ละคน
average_ratings = movie_df.groupby('Director')['Rating'].mean()

# นับจำนวนหนังที่กำกับโดยผู้กำกับแต่ละคน
movie_counts = movie_df['Director'].value_counts()

# เลือกผู้กำกับที่มี rating เฉลี่ยมากกว่า 7 และมีหนังมากกว่า 2 เรื่อง
qualified_directors = average_ratings[(average_ratings > 7) & (movie_counts >= 5)].index

# สร้าง dictionary เพื่อเก็บ genre ของผู้กำกับแต่ละคน
director_genres = {}
for director in qualified_directors:
    director_genres[director] = set(movie_df[movie_df['Director'] == director]['Genre'].str.split(', ').explode())

# เรียงลำดับผู้กำกับตาม rating เฉลี่ยจากมากไปน้อย
sorted_directors = sorted(qualified_directors, key=lambda x: average_ratings[x], reverse=True)

# แสดงผลลัพธ์
print("ผู้กำกับที่มี rating เฉลี่ยมากกว่า 7 และมีหนังไม่น้อยกว่า 5 เรื่อง (เรียงลำดับตาม rating):")
for director in sorted_directors:
    rating = average_ratings[director]
    num_movies = movie_counts[director]
    genres = ', '.join(director_genres[director])
    print(f"- {director}: ค่าเฉลี่ย {rating:.2f} จาก {num_movies} เรื่อง, ประเภท: {genres}")
````
