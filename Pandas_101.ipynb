1. นิทานที่ยาวที่สุด และสั้นที่สุด (ไม่เอาสัญลักษณ์พิเศษ)

````
folk_df = pd.read_csv(path_to_folktales)
folk_df
````

````
import pandas as pd
import re

# ฟังก์ชันสำหรับลบตัวอักษรภาษาอังกฤษและสัญลักษณ์
def remove_english_and_symbols(text):
    if isinstance(text, str): # Check if 'text' is a string
        return re.sub(r'[a-zA-Z0-9!@#$%^&*()_+=\[\]{};:"\\|,.<>\/?~`]', '', text)
    else:
        return "" # Return empty string if not a string

longest_story = ("", "", 0) # initialize longest_story (title, story, length)
shortest_story = ("", "", float('inf')) # initialize shortest_story (title, story, length)

# Assuming 'stories' is the list containing the actual stories
for index, row in folk_df.iterrows(): # Iterate over rows to access both title and description
    title = row['title']
    story = row['description']

    # กรองตัวอักษรภาษาอังกฤษและสัญลักษณ์ออก
    clean_story = remove_english_and_symbols(story)

    # คำนวณความยาวของนิทาน
    story_length = len(clean_story)

    # Update longest and shortest stories
    if story_length > longest_story[2]:
        longest_story = (title, clean_story, story_length)
    if story_length < shortest_story[2]:
        shortest_story = (title, clean_story, story_length)

# แสดงผล
print("นิทานที่ยาวที่สุด:")
print("ชื่อ:", longest_story[0])
print(longest_story[1].strip())
print("\nความยาว:", longest_story[2], "ตัวอักษร")

print("\n\nนิทานที่สั้นที่สุด:")
print("ชื่อ:", shortest_story[0])
print(shortest_story[1].strip())
print("\nความยาว:", shortest_story[2], "ตัวอักษร")
````
