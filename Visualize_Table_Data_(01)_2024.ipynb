```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name');
```

## ปรับ parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#bcd7ff', '#ffc9bc', '#ccbcff')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('c', 'm', 'y')) #ตัวย่อของสี R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((0.8,0.8,0), (0,0.7,1), (0.6,0.8,1))) #rgb value (r,g,b) | r,g,b -> [0,1]
```

## ปรับ parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='Set3')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

# Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[10,10] #กำหนดขนาดของ figure ตัวอื่นจะถูกกำหนดไปด้วย ฟีล ๆ ตั้งค่าไลบารี่นี้เลย ไม่ใช่ตั้งค่าแต่ละ plot

plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='plasma')
```

# การใช้ subplot

```
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest',cmap='magma',alpha = 0.8 ) #alpha ความเข้ม
```
