```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name');
```

## à¸›à¸£à¸±à¸š parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#bcd7ff', '#ffc9bc', '#ccbcff')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('c', 'm', 'y')) #à¸•à¸±à¸§à¸¢à¹ˆà¸­à¸‚à¸­à¸‡à¸ªà¸µ R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((0.8,0.8,0), (0,0.7,1), (0.6,0.8,1))) #rgb value (r,g,b) | r,g,b -> [0,1]
```

## à¸›à¸£à¸±à¸š parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='Set3')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

```
matplotlib.rcParams['figure.figsize']=[10,10] #à¸à¸³à¸«à¸™à¸”à¸‚à¸™à¸²à¸”à¸‚à¸­à¸‡ figure à¸•à¸±à¸§à¸­à¸·à¹ˆà¸™à¸ˆà¸°à¸–à¸¹à¸à¸à¸³à¸«à¸™à¸”à¹„à¸›à¸”à¹‰à¸§à¸¢ à¸Ÿà¸µà¸¥ à¹† à¸•à¸±à¹‰à¸‡à¸„à¹ˆà¸²à¹„à¸¥à¸šà¸²à¸£à¸µà¹ˆà¸™à¸µà¹‰à¹€à¸¥à¸¢ à¹„à¸¡à¹ˆà¹ƒà¸Šà¹ˆà¸•à¸±à¹‰à¸‡à¸„à¹ˆà¸²à¹à¸•à¹ˆà¸¥à¸° plot

plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='plasma')
```
